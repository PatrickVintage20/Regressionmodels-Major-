## Linear Regression with sci-kit-Learn API

```python
df.head()
```

```python
X = df['RM'].values.reshape(-1,1)
```

```python
y = df['MEDV'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(12,10))
sns.regplot(X,y)
plt.xlabel = ('average number of room per dwelling')
plt.ylabel = ('Median value of owner-occupied homes in $1000s')
plt.show()
```

```python
sns.jointplot(x='RM', y='MEDV', data=df, kind='reg', height=8)
```
