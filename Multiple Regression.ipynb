## Model statistics Output :

Dep Variable : The dependent Variable or Target variable

Model : Highlight the model used to obtain this output.it is OLS here. odinary least square/ Linear regression

Method : The method used to fit the data in the model.Least squares

No. observation:The number of observation

DF Residual : The degree of freedom of the residual. calculated by taking the number of observations less the number of parameter

DF Model:The number of estimated parameters in the model. in this case 13, The constant term isnt included

## Standardized Variable to Identify Key Feature(s)

- in order to perform point 2 properly, one needs to standardise the variable

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
result = pd.DataFrame(list(zip(model.coef_, df.columns )), columns = ['coefficient', 'name']).set_index('name')
np.abs(result).sort_values(by='coefficient',ascending=False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
scaler = StandardScaler()
Standard_coefficient_linear_reg = make_pipeline(scaler, model)
```

```python
Standard_coefficient_linear_reg.fit(X,y)
result = pd.DataFrame(list(zip(Standard_coefficient_linear_reg.steps[1][1].coef_, df.columns)), 
columns=['coefficient','names']).set_index('names')
np.abs(result).sort_values(by='coefficient',ascending = False)
```

## Use R^2 to identify key features

-compare R2 of model against the R2 of model without a feature

-A significant change in R2 signifies the importance of the feature

```python
from sklearn.metrics import r2_score
```

```python
linear_reg = smf.ols(formula = 'y ~ CRIM +  ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT',
                     data=df)
benchmark = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

Without LSTAT

```python
linear_reg = smf.ols(formula = 'y ~ CRIM +  ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B ',
                     data=df)
lr_without_LSTAT = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

Without AGE

```python
linear_reg = smf.ols(formula = 'y ~ CRIM +  ZN + INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT',
                     data=df)
lr_without_AGE = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
linear_reg = smf.ols(formula = 'y ~ CRIM +  ZN + INDUS + CHAS + NOX + RM + AGE + RAD + TAX + PTRATIO + B + LSTAT',
                     data=df)
Lr_without_DIS = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```
